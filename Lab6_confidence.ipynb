### Setup
For the LHC version of Lab 6, we will make fake background data in the shape of a Poisson distribution with mean 100.

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

plt.rcParams["figure.figsize"] = (16,10)

u = 100
bg = stats.poisson.rvs(mu=u, size = 1000000)
```

### Problem 1: 
The first problem is to find where the 5 sigma sensitivity threshold is. Since we know our background follows a Poisson distribution with mean (mu) = 100, we will find where the 5 sigma threshold is for that distribution.

```python
prob_5sigma = stats.norm.sf(5)
threshold = stats.poisson.isf(prob_5sigma, mu=u)
print(threshold)
```

This tells us the 5 sigma sensitivity threshold is a measurment of 154 or more. Now let's plot it to see what it looks like with our background data.

```python
plt.close()
fig, ax = plt.subplots()
ax.set_title("mu=" + str(u) + " Poisson Distribution")
ax.hist(bg, bins=range(160))
plt.tick_params(labelsize = 14)
ax.set_xlabel('number of events', size = 16)
ax.set_ylabel('Probability', size  = 16)
ax.scatter(154, 300, color="orange")
plt.show()
```

As we can see, the blue background looks like a bell-shaped normal distribution, because the Poisson distribution looks like a normal distribution when the mean is large. The orange dot is at x = 154, the 5 sigma threshold, which is higher than the background.
